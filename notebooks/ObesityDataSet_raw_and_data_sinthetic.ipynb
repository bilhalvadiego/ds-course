```python
import pandas as pd

df = pd.read_csv('data/ObesityDataSet_raw_and_data_sinthetic.csv')

df.head()
```

```python
df['CAEC'].unique()
```

```python
df['female'] = df['Gender'].map({'Male': 0, 'Female': 1})
df = df.drop(['Gender'], axis=1)
df['family_history_with_overweight_adj'] = df['family_history_with_overweight'].map({'yes': 1, 'no': 0})
df = df.drop(['family_history_with_overweight'], axis=1)
df['FAVC_adj'] = df['FAVC'].map({'yes': 1, 'no': 0})
df = df.drop(['FAVC'], axis=1)
df['CAEC_adj'] = df['CAEC'].map({'Sometimes': 1, 'Frequently': 2, 'Always': 3, 'no': 0})
df = df.drop(['CAEC'], axis=1)
df['SMOKE_adj'] = df['SMOKE'].map({'yes': 1, 'no': 0})
df = df.drop(['SMOKE'], axis=1)
df['SCC_adj'] = df['SCC'].map({'yes': 1, 'no': 0})
df = df.drop(['SCC'], axis=1)
df['CALC_adj'] = df['CALC'].map({'Sometimes': 1, 'Frequently': 2, 'Always': 3, 'no': 0})
df = df.drop(['CALC'], axis=1)

df['MTRANS_adj'] = df['MTRANS'].map({'Motorbike': 1, 'Public_Transportation': 2, 'Bike': 3, 'Walking': 4, 'Automobile': 0})
df = df.drop(['MTRANS'], axis=1)


df['NObeyesdad_adj'] = df['NObeyesdad'].map({'Insufficient_Weight':0, 'Normal_Weight': 1, 'Overweight_Level_I': 2, 'Overweight_Level_II': 3, 'Obesity_Type_I': 4, 'Obesity_Type_II': 5, 'Obesity_Type_III': 6})
df = df.drop(['NObeyesdad'], axis=1)
```

```python
df.head()
```

```python
df.describe()
```

```python
X = df.drop(['NObeyesdad_adj'], axis=1).reset_index(drop=True)
y = df['NObeyesdad_adj'].reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC

clf = SVC(kernel='linear', C=10, random_state=123)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_train_true = y_train
y_train_pred = clf.predict(X_train)
y_test_true = y_test
y_test_pred = clf.predict(X_test)


print(f"Acurácia de Treino: {round(accuracy_score(y_train_true, y_train_pred), 2)}")
print('\n ---------------------------\n')
print(f"Acurácia de Teste: {round(accuracy_score(y_test_true, y_test_pred), 2)}")
```
